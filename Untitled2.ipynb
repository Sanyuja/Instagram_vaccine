```python
# Extracting info from link list
links_df = pd.read_csv('All_Links.csv') # -- put latest file in cwd
post_data = []

for post in	links_df.loc[945:1100, '0']:
    driver.get(post)
    #time.sleep(6)
    #Username
    try: 
        username = driver.find_element_by_xpath('//*[@id="react-root"]/section/main/div/div[1]/article/header/div[2]/div[1]/div[1]/span/a').text
    except :
        username = np.nan
        pass
    

    #post_image -- this targets all images on the page but ultimately keeps the one we want
    try:
        pics = driver.find_elements_by_tag_name('img')
        imgs = [image.get_attribute('src') for image in pics]
        post_image = imgs[1:2] #first img on post pages are user's profile, second is img relating to actual post -- later images are for profiles of users that comment
        desc = [image.get_attribute('alt') for image in pics]
        desc = desc[1]
    except:
        post_image='vid'
        pass
        

    #post caption
    try:
        caption = driver.find_element_by_xpath('//*[@id="react-root"]/section/main/div/div[1]/article/div[3]/div[1]/ul/div/li/div/div/div[2]/span').text
    except :
        caption = np.nan
    
    #post_date
    try:
        post_date = driver.find_element_by_css_selector('/html/body/div[5]/div[2]/div/article/div[3]/div[2]/a/time').text
    except :
        post_date = np.nan


    #post num of likes or views
    try:
        likes = driver.find_element_by_xpath('//*[@id="react-root"]/section/main/div/div[1]/article/div[3]/section[2]/div/div/a/span').text
    except :
        likes = np.nan
        pass

    try:
        views = driver.find_element_by_xpath('/html/body/div[5]/div[2]/div/article/div[3]/section[2]/div/span/span').text
    except :
        views = np.nan
        pass


    #comments
    try:
        comment_1 = driver.find_element_by_xpath('/html/body/div[1]/section/main/div/div[1]/article/div[3]/div[1]/ul/div/li').text
    except :
        comment_1 = np.nan
        pass

    try:
        comment_2 = driver.find_element_by_xpath('/html/body/div[1]/section/main/div/div[1]/article/div[3]/div[1]/ul/ul/div/li').text
    except :
        comment_2 = np.nan
        pass


    #location
    try:
        xpath_location = '/html/body/div[1]/section/main/div/div[1]/article/header/div[2]/div[2]'
        location = driver.find_element_by_xpath(xpath_location).text
    except :
        location = np.nan
        pass

    
    
    # Grab User Information
    driver.get(f'https://www.instagram.com/{username}/')


    #bio
    try:
        bio = driver.find_element_by_xpath("/html/body/div[1]/section/main/div/div[1]").text
    except :
        bio = np.nan
        pass

    #following & followers
    try:
        following = driver.find_element_by_xpath('/html/body/div[1]/section/main/div/header/section/ul/li[3]/a').text
    except :
        following = np.nan
        pass

    try:
        followers = driver.find_element_by_xpath('/html/body/div[1]/section/main/div/header/section/ul/li[2]/a').text
    except :
        followers = np.nan
        pass


    # Append information to pd df -- will create csv for storage later!
    post_data.append([username, post_image, desc,caption, post_date, likes, views, comment_1, comment_2, location, bio, following, followers])
    

# once done with scraping/appending info from all posts -- concat into df & save to csv
posts_df = pd.DataFrame(post_data, columns=['username', 'post_image','descrip', 'caption', 'post_date', 'likes', 'views', 'comment_1', 'comment_2', 'location', 'bio', 'following', 'followers'])
#posts_df.to_csv(f'posts_data -- size{len(posts_df)}.csv', index=False)

```
